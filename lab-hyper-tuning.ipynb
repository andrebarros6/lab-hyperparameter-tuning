```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

Now perform the same as before:
- Feature Scaling
- Feature Selection


```python
spaceship.isna().sum()
spaceship = spaceship.dropna()
```

```python
#split by '/' 
spaceship[['Cabin', 'Deck', 'Side']] = spaceship["Cabin"].str.split("/", expand = True)
```

```python
spaceship.drop(columns = ["PassengerId", "Name"], inplace = True)
```

```python
# label HomePlanet
# Earth = 0
# Mars = 1
# Europa = 2

def label_planets(planet):
        if planet == "Earth":
                return 0
        elif planet == "Mars":
                return 1
        else:
                return 2

spaceship["HomePlanet"] = spaceship["HomePlanet"].apply(label_planets)
```

```python
#label CryoSleep 
# False = 0
# True = 1

def label_cryo(cryosleep):
        if cryosleep == False:
                return 0
        elif cryosleep == True:
                return 1
        
spaceship["CryoSleep"] = spaceship["CryoSleep"].apply(label_cryo)
```

```python
# label Cabin
# A = 0
# B = 1
# C = 2
# ...
#  T = 7

def label_cabin(cabin):
        if cabin == "A":
                return 0
        elif cabin == "B":
                return 1
        elif cabin == "C":
                return 2
        elif cabin == "D":
                return 3
        elif cabin == "E":
                return 4
        elif cabin == "F":
                return 5
        elif cabin == "G":
                return 6
        else:
                return 7

spaceship["Cabin"] = spaceship["Cabin"].apply(label_cabin)
```

```python
# label Destination
# TRAPPIST-1e = 10
# PSO J318.5-22 = 11
# 55 Cancri e = 12

def label_destiny(planet):
        if planet == "TRAPPIST-1e":
                return 10
        elif planet == "PSO J318.5-22":
                return 11
        else:
                return 12

spaceship["Destination"] = spaceship["Destination"].apply(label_destiny)
```

```python
# label Side
# S = 0
# P = 1

def label_side(side):
        if side == "S":
                return 0
        else:
                return 1

spaceship["Side"] = spaceship["Side"].apply(label_side)
```

- Now let's use the best model we got so far in order to see how it can improve when we fine tune it's hyperparameters.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.metrics import accuracy_score 
```

```python
#adaBoost is the model that has produced the best scores so far
```

```python
# split df in features and target
features = spaceship.drop("Transported", axis=1)
target = spaceship["Transported"]
```

```python
#split into test and train data
X_train, X_test, y_train, y_test =  train_test_split(features, target, test_size=0.20)
```

```python
#creating the normalizer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
normalizer = StandardScaler()

normalizer.fit(X_train)
```

```python
#transform training and testing data - needed to transform both so the results are comparable
X_train_norm = normalizer.transform(X_train)
X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_train_norm.head()
```

```python
ada_class = AdaBoostClassifier(DecisionTreeClassifier(max_depth=6),
                            n_estimators=15)
ada_class.fit(X_train_norm, y_train)
```

- Evaluate your model

```python
pred = ada_class.predict(X_test_norm)
accuracy_score(y_test, pred)
```

```python
pred = ada_class.predict(X_train_norm)
accuracy_score(y_train, pred)
```

**Grid/Random Search**

For this lab we will use Grid Search.

- Define hyperparameters to fine tune.

```python
grid = {"n_estimators": [6, 8, 10, 12, 15],
        "estimator__max_leaf_nodes": [250, 300, 600, 1000, 2000],
        "estimator__max_depth":[6, 8, 12, 15]}
```

- Run Grid Search

```python
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(estimator = ada_class, param_grid = grid, cv=5, verbose=5, return_train_score=True)
```

```python
model.fit(X_train_norm, y_train)
```

- Evaluate your model

```python
model.best_params_
```

```python
best_model = model.best_estimator_
```

```python
pred = best_model.predict(X_test_norm)
accuracy_score(y_test, pred)
```

```python
pred = best_model.predict(X_train_norm)
accuracy_score(y_train, pred)
```
